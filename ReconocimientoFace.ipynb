```python
!pip install deepface

import urllib.request
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Descargar la imagen
img_url = 'https://img.freepik.com/foto-gratis/comunidad-jovenes-positivos-juntos_23-2148431352.jpg'
img_path = 'personas.jpg'
urllib.request.urlretrieve(img_url, img_path)

# Analizar la imagen
analysis = DeepFace.analyze(img_path, actions=['gender', 'age', 'race', 'emotion'], enforce_detection=False)

data = []

# Función para clasificar la edad
def classify_age(age):
    return 'Adulto' if age >= 18 else 'Niño'


# Leer la imagen y convertir a RGB
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Analizar cada persona detectada
for person in analysis:
    gender = person['dominant_gender']
    age = person['age']
    race = person['dominant_race']
    emotion = person['dominant_emotion']
    region = person['region']
    x, y, w, h = region['x'], region['y'], region['w'], region['h']

    # Clasificar grupo de edad
    age_group = classify_age(age)

    # Almacenar la información
    data.append({
        'Género': gender,
        'Edad': age,
        'Raza': race,
        'Emoción': emotion,
        'Posición': f"({x}, {y}), Ancho: {w}, Alto: {h}",
        'Edad Grupo': age_group
    })

    # Dibujar el rectángulo y el texto en la imagen
    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(image_rgb, gender, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)


# Crear DataFrame con los datos
df = pd.DataFrame(data)

# Mostrar los gráficos
plt.imshow(image_rgb)
plt.axis('off')
plt.title(f"Detectadas {len(analysis)} caras")
plt.show()

# Mostrar la tabla debajo de los gráficos
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('tight')
ax.axis('off')

# Crear la tabla
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Ajustar el tamaño de la tabla
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

plt.show()

# Mostrar las distribuciones
gender_counts = df['Género'].value_counts()
age_group_counts = df['Edad Grupo'].value_counts()
race_counts = df['Raza'].value_counts()

# Gráfico de distribución de género
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
plt.title('Distribución de Género')
plt.show()

# Gráfico de distribución de edad (Adulto vs Niño)
plt.figure(figsize=(8, 6))
age_group_counts.plot(kind='bar', color=['lightblue', 'lightgreen'])
plt.title('Distribución de Edad (Adulto vs Niño)')
plt.xlabel('Grupo de Edad')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()

# Gráfico de distribución de raza
plt.figure(figsize=(8, 6))
race_counts.plot(kind='bar', color=['orange', 'yellow', 'green', 'blue'])
plt.title('Distribución de Raza')
plt.xlabel('Raza')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()








```
